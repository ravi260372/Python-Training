## Segmentation Targeting & Positioning

**Segmentation**   
-- The process of dividing a population of customers into groups that share similar characteristics.  
-- Observations within the same group would have comparable purchasing behavior.  
-- Observations within the same group would respond similarly to different marketing activities.  

*Why Segmentation* : Individuals within segments are expected to have similar purchase behaviors and respond similarly to different marketing activities. Segmentation helps marketers to better understand customer behavior and develop better ways to offer the value proposition of the company  

*Consumer Data used by Marketing*  
-- Demographic (Age , Gender , Education)    
-- Geographic  
-- Psychographic (Planned Buyer , Impulsive Buyer)   
-- Behavioural (Existing Historical Data : Purchase Frequency , Time of Purchase , Quantity Purchased , Product Ratings)  

**Targeting** *More into the advertising territory*    
-- The process of evaluating potential profits from each segment and deciding which segments to focus on.  
---- *Take into account Segment Size , Expected Growth , Competitor's Offering*   
-- Selecting ways to promote your products. You can target one segment on TV and another online.  
-- Examining customersâ€™perception (Involves psychology and usually budget constraints)    

**Positioning**   
-- What product would have the closest characteristics to what the customers' needs(What product can be offered)   
-- What product characteristics do the customers from a certain segment need?  
-- This mainly concerns implementing the "Targeting Strategies" i.e. How it should be presented to cutomers and through what channel   
*What is positioning* :   
(1) The characteristics a product should have in order to satisfy customer needs;  
(2) How the product should be presented to customers;  
(3) Through what channel the product should be presented to customers;  

## Marketing Mix  

**Develop the best product or service and offer it at the right price through the right channels**    
Develop Models that estimate the following : 
(1) Purchasing Probability   
(2) Brand Choice Probability  
(3) Purchase Quantity    

**4Ps of Marketing Mix**  
(1) Product Characteristics : Core Attributes of Product : Product Features; Branding; Packaging    

(2) Price of Offering : Product Cost; Long term price changes ; Discounts     

(3) Promotion : How Product is Communicated / Advertised & through which channel : *Sales Promotion / Merchandising is done through Price reduction, Display (location) and Feature (Printed Ads , Hand Outs , News Paper inserts)*    

(4) Place / Channel : Exact location where the Product will be offered / Distributed : *Distribution Strategies* : Intensive (Many Locations) , Selective (Specific Locations with best selling chance) , Exclusive (Single Branded Location e.g. Luxury Goods)

## Price Elasticity  

Price elasticity measures how a variable of interest (e.g Purchasing Behaviour) changes when the price changes. *Example -- Price Increase of INR 10 for Coca Cola results in reduced purchases i.e. lesser people buying alternatively Price Reduction of INR 5 for Coca Cola results in higher purchases i.e. more people buying*    

 

Three types of Price Elasticity  


Aggregate Price of All Types of Beer Increases from USD 1.5 to USD 2.5 implies reduced Purchase Probability. Calculating Price Elaticity of Purchase Price would show how much Purchase Quantity will decrease    


Price of Oreo Cookies increases compared to other brands implying Sales of Oreo Decrease. Calculating Price Elasticity of Brand Choice will show how much purchase quantity for a particular brand will change with Price Change of that Brand.
