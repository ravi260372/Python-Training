# Introduction

A DataFrame is an object that stores data as rows and columns.  

You can think of a DataFrame as a spreadsheet or as a SQL table.   

You can manually create a DataFrame or fill it with data from a CSV, an Excel spreadsheet, or a SQL query.

**Each column has a name, which is a string.**  
**Each row has an index, which is an integer.** 

DataFrames can contain many different data types: strings, ints, floats, tuples, etc.

# Creating a DataFrame

## Creating DataFrame : passing a Dictionary  

You can pass in a dictionary to **pd.DataFrame(**).   

*Each key is a column name and each value is a list of column values.* The columns must all be the same length or you will get an error

You need to import the "pandas library" to create and manipulate DataFrames  

The leftmost column shown when we print out a dataframe is known as the “index”, or “row labels”, of a dataframe.

By default, every dataframe that you create or load from a file will have this column of values which starts from 0 and increases sequentially for each row. These values are essentially like the identification for each row, as each value should be unique per row.

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'name': ['John Smith', 'Jane Doe', 'Joe Schmo'],
    'address': ['123 Main St.', '456 Maple Ave.', '789 Broadway'],
    'age': [34, 28, 51]
})
df1
```

## Creating DataFrame : using Lists  
You can also add data using lists.

For example, you can pass in a list of lists, where each list represents a row of data.   

We use the keyword argument **columns** to pass a list of column names. 

```python
df2 = pd.DataFrame([
    ['John Smith', '123 Main St.', 34],
    ['Jane Doe', '456 Maple Ave.', 28],
    ['Joe Schmo', '789 Broadway', 51]
    ],
    columns=['name', 'address', 'age'])
df2
```

**Question :** When should I create a Pandas dataframe using a dictionary or a list?  
**Answer :** You can create Pandas dataframes using either a dictionary or a list of lists, but depending on several factors, using one can be preferred over the other.

Using dictionaries can be much faster, since you can just include the column names as the keys, and include the values of the column as a list for the keys. However, a disadvantage of using a dictionary is that the columns will not preserve the order that you entered them, and will default to alphabetical ordering instead. This is important to keep in mind especially if the column order is important.

Using a list of lists allows you to enter each row of data one at a time as a separate list, but it may take longer than using a dictionary, since column names must be added as a separate list after the rows are added. However, a list of lists allows you to order the column names specifically, which can be very important.

**Question :** How is a Pandas series different from a dataframe?  
**Answer :** 
* In Pandas a series is a one-dimensional object that contains any type of data, similar in ways to a Numpy array.

* Series objects have a single axis label, like a column title, which is the index of the series. A series is essentially a single column  

* A dataframe is a two-dimensional object that can hold multiple columns of different types of data. They are similar to a table in SQL.

* A single column of a dataframe is a series, and a dataframe is a container of two or more series objects

```python
# Creating a series
clinic_east = pd.Series([100, 51, 81, 80, 51, 112])
print(clinic_east)
print(type(clinic_east))
```

```python
# Creating a DataFrame
df = pd.DataFrame ([
  ['January', 100, 100],
  ['February', 51, 45],
  ['March', 81, 96]],
  columns=["month", "clinic_east", "clinic_north"]
)
print(df)
print(type(df))
```

## Lambda Function: If Statements

We can make our lambdas more complex by using a modified form of an if statement.

**Syntax : lambda x: [OUTCOME IF TRUE] if [CONDITIONAL] else [OUTCOME IF FALSE]**

Suppose we want to pay workers time-and-a-half for overtime (any work above 40 hours per week). 

The following function will convert the number of hours into time-and-a-half hours using an if statement  
We can accomplish the same with a lambda function

```python
# function approach
def myfunction(x):
    if x > 40: # if work hours are > 40 
        return 40 + (x - 40) * 1.50 # apply overtime rate for hours > 40 + base hours
    else:
        return x # if work hours <= 40 then normal
# lambda approach
myfunction_lambda = lambda x: 40 + (x - 40) * 1.50 if x > 40 else x

print("Without Lambda Function")
print(myfunction(45))
print("With Lambda Function")
print(myfunction_lambda(45))
```

## Renaming Columns : ".rename" method  
You also can rename individual columns by using the .rename method. Pass a dictionary as below  
**{'old_column_name1': 'new_column_name1',   
'old_column_name2': 'new_column_name2'}**

```python
# creat a dataframe
df = pd.DataFrame({
    'name': ['John', 'Jane', 'Sue', 'Fred'],
    'age': [23, 29, 21, 18]
})

# print(df)
print("original DataFrame :")
print(df)

# rename the columns
df.rename(columns={
    'name': 'First Name',
    'age': 'Age'},
    inplace=True)

print("new DataFrame :")
print(df)
```

The code above will rename **name to First Name** and **age to Age.**

**Using rename with only the columns keyword will create a new DataFrame, leaving your original DataFrame unchanged.**  

That’s why we also passed in the keyword argument inplace=True. 

Using inplace=True lets us edit the original DataFrame.

There are several reasons why .rename is preferable to .columns:

    You can rename just one column
    You can be specific about which column names are getting changed 
    With .column you can accidentally switch column names if you’re not careful

*Note: If you misspell one of the original column names, this command won’t fail. It just won’t change anything.*

If we didn’t know that 'movies' was a table of movie ratings, the column name might be confusing.

1. column1 is the column that we want to group by
2. column2 is the column that we want to perform a measurement on
3. measurement is the measurement function we want to apply

**The output of a "groupby" is a "series object"**

## Aggregating and Re-setting Index (transform to DataFrame)
After using groupby, we often need to clean our resulting data.

The groupby function creates a new Series, not a DataFrame. The indices of the Series have different values as compared to the indices in the original DataFrame.

Usually, we’d prefer that those indices were actually a column.   
In order to get that, we can use reset_index(). This will transform our Series into a DataFrame and move the indices into their own column.  

Generally, you’ll always see a groupby statement followed by reset_index:  
**df.groupby('column1').column2.measurement().reset_index()**  

When we use groupby, we often want to rename the column we get as a result..   
**new_df = df.groupby('column1').column2.measurement().reset_index() : new DataFrame**  
**new_df = new_df.rename(columns = {"old_col_name":"new_col_name"} : rename the new column**

# Aggregating in Pandas : Pivot Tables  
When we perform a groupby across multiple columns, we often want to change how our data is stored. We can achieve this by implementing the "Pivot Table" functionality from "MSEXCEL" in Pandas...

**What is a Pivot Table**  
A pivot table is a table of statistics that summarizes the data of a more extensive table (such as from a database, spreadsheet, or business intelligence program). This summary might include sums, averages, or other statistics, which the pivot table groups together in a meaningful way.
Pivot tables are a technique in data processing. They arrange and rearrange (or "pivot") statistics in order to draw attention to useful information.   

**Syntax for Pivot Table**  
df.pivot(columns='ColumnToPivot',  
.........index='ColumnToBeRows',  
.........values='ColumnToBeValues')

# Project : A/B Testing for ShoeFly.com  
Our favorite online shoe store, ShoeFly.com is performing an A/B Test.   

They have two different versions of an ad, which they have placed in emails, as well as in banner ads on Facebook, Twitter, and Google.   

They want to know how the two ads are performing on each of the different platforms on each day of the week.   

Help them analyze the data using aggregate measures.

## Analyzing Add Sources

```python
## loading libraries
import pandas as pd
import numpy as np

## loading the data
ad_clicks = pd.read_csv('add_clicks.csv')

## inspecting top 5 rows
ad_clicks.head()
```

Your manager wants to know which ad platform is getting you the most views.

How many views (i.e., rows of the table) came from each utm_source?


```python
# grouping ad_clicks by source and counting user_ids
ad_clicks.groupby('utm_source')\
.user_id.count()\
.reset_index()
```

If the column ad_click_timestamp is not null, then someone actually clicked on the ad that was displayed.

Create a new column called is_click, which is True if ad_click_timestamp is not null and False otherwise.

```python
# use the "isnull() method on "time-Stamp" column
ad_clicks['is_click'] = ~ad_clicks\
   .ad_click_timestamp.isnull()
ad_clicks.head()

# The ~ is a NOT operator, and isnull() tests whether or not the value of ad_click_timestamp is null.
# isnull() method will return "True" if value is present else will return "False"
```

We want to know the percent of people who clicked on ads from each utm_source.

Start by grouping by utm_source and is_click and counting the number of user_id‘s in each of those groups.   

Save your answer to the variable clicks_by_source

```python
clicks_by_source = ad_clicks\
.groupby(['utm_source' , 'is_click'])\
.user_id.count()\
.reset_index() 

clicks_by_source
```

Now let’s pivot the data so that the columns are is_click (either True or False), the index is utm_source, and the values are user_id.

Save your results to the variable clicks_pivot.


```python
clicks_pivot = clicks_by_source.pivot(
columns = 'is_click',
index = 'utm_source',
values = 'user_id')\
.reset_index()

clicks_pivot
```

Create a new column in clicks_pivot called percent_clicked which is equal to the percent of users who clicked on the ad from each utm_source.

Was there a difference in click rates for each source?


```python
clicks_pivot['percent_click'] = (clicks_pivot[True]/(clicks_pivot[True] + clicks_pivot[False]))*100

clicks_pivot

# clicks_pivot[True] is the number of people who clicked (because is_click was True for those users)

# clicks_pivot[False] is the number of people who did not click (because is_click was False for those users)

# So, the percent of people who clicked would be (Total Who Clicked) / (Total Who Clicked + Total Who Did Not Click)

# finally we multiply by 100 to give value in %
```

## Analyzing an A/B Test

The column experimental_group tells us whether the user was shown Ad A or Ad B.

Were approximately the same number of people shown both adds?


```python
ad_clicks.head()
```

```python
# group by experimental group
ad_clicks.groupby('experimental_group').user_id.count().reset_index()

## equal number of users have been selected for tirals i.e. 827 per group ##
```

Using the column is_click that we defined earlier, check to see if a greater percentage of users clicked on Ad A or Ad B.

```python
# create a grouped data frame , grouped by experimental group and is-click
add_click_groups_clicks = ad_clicks.groupby(['experimental_group' , 'is_click']).user_id.count().reset_index()
add_click_groups_clicks
```

```python
# create a pivot table indexed on experimental group by is_click
add_clicks_A_B_pivot = add_click_groups_clicks.pivot(
columns = 'is_click',
index = 'experimental_group' , 
values = 'user_id').reset_index()

# view the new pivot data
add_clicks_A_B_pivot
```

```python
# calculate % of experimental groups where "is_click = True"
add_clicks_A_B_pivot['percent_click'] = \
(add_clicks_A_B_pivot[True]/(add_clicks_A_B_pivot[True]\
                             + add_clicks_A_B_pivot[False]))*100

add_clicks_A_B_pivot

# add_clicks_A_B_pivot[True] is the number of people who clicked (because is_click was True for those users)

# add_clicks_A_B_pivot[False] is the number of people who did not click (because is_click was False for those users)

# So, the percent of people who clicked would be (Total Who Clicked) / (Total Who Clicked + Total Who Did Not Click)

# finally we multiply by 100 to give value in %

### A greater percentage 37.48% users clicked on advertisement A whilst 30.83% users clicked on advertisement B ###
```

The Product Manager for the A/B test thinks that the clicks might have changed by day of the week.

Start by creating two DataFrames: a_clicks and b_clicks, which contain only the results for A group and B group, respectively.


```python
# create a DataFrame with "A Group Users"
a_clicks = ad_clicks[
   ad_clicks.experimental_group
   == 'A']

# create a DataFrame with "B Group Users"
b_clicks = ad_clicks[
   ad_clicks.experimental_group
   == 'B']
```

```python
a_clicks.head(2)
```

```python
b_clicks.head(2)
```

For each group (a_clicks and b_clicks), calculate the percent of users who clicked on the ad by day

```python
## create a grouped DataFrame of A& B users grouped by is_click and day and count the number of users
a_clicks_group = a_clicks.groupby(['is_click' , 'day']).user_id.count().reset_index()

b_clicks_group = b_clicks.groupby(['is_click' , 'day']).user_id.count().reset_index()
```

```python
# pivot a_click DataFrame indexed by "Day" with columns as "is_click"
a_click_pivot = a_clicks_group.pivot(
columns = 'is_click',
index = 'day' , 
values = 'user_id').reset_index()

a_click_pivot
```

```python
# pivot b_click DataFrame indexed by "Day" with columns as "is_click"
b_click_pivot = b_clicks_group.pivot(
columns = 'is_click',
index = 'day' , 
values = 'user_id').reset_index()

b_click_pivot
```

```python
# calculate % of clicks for each day for each group A
a_click_pivot['percent_click_a'] = \
(a_click_pivot[True]/(a_click_pivot[True]\
                             + a_click_pivot[False]))*100

a_click_pivot
```

```python
# calculate % of clicks for each day for each group B
b_click_pivot['percent_click_b'] = \
(b_click_pivot[True]/(b_click_pivot[True]\
                             + b_click_pivot[False]))*100

b_click_pivot
```

```python
# create a final data frame with values of "click%" from both clickA and clickB

final_df = pd.merge(a_click_pivot[['day','percent_click_a']], b_click_pivot[['day','percent_click_b']], on='day')


# add a column of differences in click % (A-B)
final_df['difference'] = final_df['percent_click_a'] - final_df['percent_click_b']

# view the final table
final_df
```
